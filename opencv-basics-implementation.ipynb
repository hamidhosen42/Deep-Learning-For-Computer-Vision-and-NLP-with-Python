**Image Masking with OpenCV**

We can apply a mask to an image by computing the cv2.bitwise_and() between the mask and the image. To track a color, we define a mask in HSV color space using cv2.inRange() passing lower and upper limits of color values in HSV.

To track a part of the image we can define a mask using np.zeros() and slicing the entries with white (255) for the region in the input image to examine. Follow the given steps to mask an image âˆ’

The first step is to import required libraries. The required Python libraries are OpenCV, and NumPy. Make sure you have already installed them.

Next read the input image using cv2.imread() method. Convert the image BGR to HSV to track a color in the input image. To track a part of the image leave the image in BGR format.

Define a mask using cv2.inRange() to track a particular color in the image. Pass the lower and upper limits of color in HSV format. If you want to track the rectangular part of the input image, define a rectangular image (mask) with cv2.zeros(). Fill the entries of the mask with 255 to track the area in the original image.

Apply bitwise AND operation between the mask and the input image using cv2.bitwise_and().

```python
#mask
Image_mask = np.zeros(imagedata.shape[:2],dtype = 'uint8') 
```

**OpenCV - Rotation**

You can perform rotation operation on an image using the warpAffine() method of the imgproc class. Following is the syntax of this method âˆ’

Imgproc.warpAffine(src, dst, rotationMatrix, size);

This method accepts the following parameters âˆ’

src âˆ’ A Mat object representing the source (input image) for this operation.

dst âˆ’ A Mat object representing the destination (output image) for this operation.

rotationMatrix âˆ’ A Mat object representing the rotation matrix.

size âˆ’ A variable of the type integer representing the size of the output image.

```python
# Extracting height and width from 
# image shape
(imageH,imageW) = imagedata.shape[:2]
(imageH,imageW)
```

```python
# get the center coordinates of the
# image to create the 2D rotation
# matrix
centerX, centerY = (imageH//2,imageW//2)
(centerX,centerY)
```

```python
# using cv2.getRotationMatrix2D() 
# to get the rotation matrix
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),45,1.0) #center, angle, scale

# rotate the image using cv2.warpAffine 
# 45 degree anticlockwise
rotateNow = cv2.warpAffine(imagedata, imageRotate, (imageH, imageW))
catimageShow("Rotated image",rotateNow)
```

```python
# using cv2.getRotationMatrix2D() 
# to get the rotation matrix
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),90,1.0) #center, angle, scale

# rotate the image using cv2.warpAffine 
# 90 degree anticlockwise
rotateNow = cv2.warpAffine(imagedata, imageRotate, (imageH, imageW))
catimageShow("Rotated image",rotateNow)
```

```python
# using cv2.getRotationMatrix2D() 
# to get the rotation matrix
imageRotate = cv2.getRotationMatrix2D((centerX,centerY),180,1.0) #center, angle, scale

# rotate the image using cv2.warpAffine 
# 180 degree anticlockwise
rotateNow = cv2.warpAffine(imagedata, imageRotate, (imageH, imageW))
catimageShow("Rotated image",rotateNow)
```

**OpenCV - Colored Images to GrayScale**

This method accepts the following parameters âˆ’

src âˆ’ A matrix representing the source.

**Sobel operation**

1. Convert the image to grayscale
2. Define the Sobel kernels
3. Edge Detection

```python
kernelGen = -1 if argmumentImage['scharr'] > 0 else 3 # > x,< y
kernelGen
```

```python
gradienImageDataX = cv2.Sobel(grayimage,ddepth = cv2.CV_32F,dx = 1, dy = 0, ksize = kernelGen) #Sobel(src, ddepth, dx, dy[, dst[, ksize[, scale[, delta[, borderType]]]]])
gradienImageDataY = cv2.Sobel(grayimage,ddepth = cv2.CV_32F,dx = 0, dy = 1, ksize = kernelGen) #Sobel(src, ddepth, dx, dy[, dst[, ksize[, scale[, delta[, borderType]]]]])

catimageShow("Gradient Image",gradienImageDataX)
```
